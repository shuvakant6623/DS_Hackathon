```python
import json
import pandas as pd
import csv
```

```python
# Input JSON file
input_path = "../Datasets/raw/conversational_transcript_dataset.json"

# Output CSV file
output_csv_path = "../Datasets/processed/transcript_dataset.csv"
output_json_path = "../Datasets/processed/conversation_transcript_map.json"

with open(input_path, 'r') as f:
    data = json.load(f)

with open(output_csv_path, 'w',newline="") as csvFile:
    fieldNames = ['transcript_id','time_of_interaction','domain','intent','reason_for_call']
    writer = csv.DictWriter(csvFile, fieldnames=fieldNames)
    writer.writeheader()

    for transcript in data["transcripts"]:
        writer.writerow({
            "transcript_id": transcript["transcript_id"],
            "time_of_interaction": transcript["time_of_interaction"],
            "domain": transcript["domain"],
            "intent": transcript["intent"],
            "reason_for_call": transcript["reason_for_call"]
        })
# Save the processed data as JSON

df = pd.read_csv(output_csv_path)
df.head(5)
```

```python
conversations_map = {}

for transcript in data["transcripts"]:
    conversations_map[transcript["transcript_id"]] = transcript["conversation"]

with open(output_json_path, "w", encoding="utf-8") as f:
    json.dump(conversations_map, f, indent=2, ensure_ascii=False)
```
